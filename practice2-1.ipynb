```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Number_of_tigers = np.array([60,350,400,2500,500,550])
types=["south china","indo_chinese","siberian","malayan","bengal","sumatran"]
```

```python

mycolors=["purple","blue","green","yellow","orange", "red"]
plt.pie(Number_of_tigers,colors=mycolors, startangle=90)
plt.legend(title="Tigers:",labels=types)
plt.title("Tigers in the wild")
```
