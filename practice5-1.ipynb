```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('summer.csv')
df
```

```python
#How many cities have the Summer Olympics been held?

Cities= df['City'].unique()
total_city =len(Cities)
total_city 

data = []

for City in Cities:
    host_count = 0
    for year in df['Year'].unique():
        if len(df[(df['City'] == City) & (df['Year'] == year)]) > 0:
          host_count += 1
    data.append([City, host_count])
    
plot_data = pd.DataFrame(data, columns = ['City', 'No. of times hosted']).sort_values(by = 'No. of times hosted', ascending= False )
plot_data.plot(x = 'City', y = 'No. of times hosted', kind='bar', figsize=(15, 5), color='#D8BFD8')
```

```python
#Which sport has won the most gold medals?

gold = df[df['Medal'] == 'Gold']

data = []

for sport in gold['Sport'].unique():
    data.append([sport, len(gold[gold['Sport'] == sport])])
    
data = pd.DataFrame(data, columns=['Sport', 'Gold']).sort_values(by= 'Gold', ascending=False).head()
data.plot(x= 'Sport', y= 'Gold', kind='bar', figsize=(10, 5), color = '#FFE4E1' )
```

```python
#Which sport has won the most medals?

groupedsport = df.groupby('Sport').count()['Medal']
groupedsport.sort_values(ascending=False).head().plot.bar(figsize=(10, 5), color='#FFDAB9')
```

```python
#Which player has won the most medals?

groupedplayer = df.groupby('Athlete').count()['Medal']
groupedplayer.sort_values(ascending=False).head().plot.bar(figsize=(10,5), color='#E6E6FA')

```

```python
#Which player has won the most gold medals?

gold = df[df['Medal'] == 'Gold']

groupedgold = gold.groupby('Athlete').count()['Medal']
groupedgold.sort_values(ascending=False).head().plot.bar(figsize=(10,5) , color='skyblue')

```

```python
#In which year did India win its first gold medal at the Summer Olympics?

gold = df[(df['Medal'] == 'Gold') & (df['Country'] == 'IND')]
year = gold['Year'].min()
year
```

```python
#Which competition has the most participants and is the most popular?

groupeddf = df.groupby('Event').count()['Athlete']
groupeddf.sort_values(ascending=False).head().plot.bar(figsize=(10,5), color='Pink')
```

```python
#Which sport has the most female gold medal winners?

gold_female = df[(df['Medal'] == 'Gold') & (df['Gender'] == 'Women')]
data = []


for Sport in gold_female['Sport'].unique():
    gold_count = gold_female[gold_female['Sport'] == Sport].shape[0]
    data.append([Sport, gold_count])
    
plot_data = pd.DataFrame(data, columns = ['Sport', 'No. of Gold medals']).sort_values(by = 'No. of Gold medals', ascending= False )
plot_data.plot(x = 'Sport', y = 'No. of Gold medals', kind='bar', figsize=(15, 5), color='#D8BFD8')

    

```
