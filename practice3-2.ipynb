```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 6, 10, 11])
y = np.array([1.5, 5, 6,5.5])

```

```python

z = np.array([2, 8, 13, 15])
j = np.array([2, 7, 9,10.5])
```

```python
plt.plot(x, y, "*", ls = "dashed", color = "red")
plt.plot(z, j, "*", ls = "dashed", color = "orange")
plt.xlabel("AGE IN WEEKS")
plt.ylabel("WEIGHT IN KILOGRAM")
plt.grid()
plt.show()
```
