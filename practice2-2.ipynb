```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x= np.array([0,6,10])
y=np.array([1.5,5,6])
z=np.array([1.5,7,9])
x_axis=np.arange(len(x))
plt.bar(x_axis-0.2, y, 0.4, label="T.J", color="purple")
plt.bar(x_axis+0.2, z, 0.4, label="Mattew", color="pink")
plt.xticks(x_axis,x)
plt.xlabel("AGE IN WEEKS")
plt.ylabel("WEIGHT IN KILOGRAM")
plt.title("TIGERS WEIGHT")
plt.legend()
```
